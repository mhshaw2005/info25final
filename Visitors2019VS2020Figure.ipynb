```python
# Imports
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Getting the data
n = open("Data\Park_visitor_data_1.json","r")
data = json.loads(n.read())
df = pd.DataFrame(data['parks'])
```

```python
#Subsetting the data frame to get the visitor numbers and park names
mapDf = df[['name','visitor_history']].copy()
mapDf.head()
```

```python
#Initializing a dictionary for each year of interest
dict2019 = {}
dict2020 = {}
dict2021 = {}

#Extracting visitor numbers and correlating index for each interest year
for i in (mapDf['visitor_history']):
    for j in i:
        if (j['year']=='2019'):
            index=i.index(j)
            dict2019[index]=j['visitors']

for i in (mapDf['visitor_history']):
    for j in i:
        if (j['year']=='2020'):
            index=i.index(j)
            dict2020[index]=j['visitors']


for i in (mapDf['visitor_history']):
    for j in i:
        if (j['year']=='2021'):
            index=i.index(j)
            dict2021[index]=j['visitors']
```

```python
#2020 Graph
names = []
number= list(dict2020.keys())
#Using the index values to extract the park's name for each population value
for i in number:
    names.append(mapDf['name'][i])
counts = list(dict2020.values())
#Creating and editing the actual plot
plt.bar(names,counts)
plt.xticks(names, rotation=90)
plt.ylabel('Number of Visitors (1e6)')
plt.title('2020 Park Visitors')
plt.savefig('2020Barplot.png')
```

```python
#Same procedure for 2019
names = []
number= list(dict2019.keys())
for i in number:
    names.append(mapDf['name'][i])
counts = list(dict2019.values())
plt.bar(names,counts)
plt.xticks(names, rotation=90)
plt.ylabel('Number of Visitors (1e6)')
plt.title('2019 Park Visitors')
plt.savefig('2019Barplot.png')
```

```python
#Same procedure for 2021
names = []
number= list(dict2021.keys())
for i in number:
    names.append(mapDf['name'][i])
counts = list(dict2021.values())
plt.bar(names,counts)
plt.xticks(names, rotation=90)
plt.ylabel('Number of Visitors (1e6)')
plt.title('2021 Park Visitors')
plt.savefig('2021Barplot.png')
```
